## 2-1. 외부 파일 읽어오기

#### 예제 2-1 csv파일 읽기

pd.read_csv(file, ...)

* header : '열 이름' 이 되는 행 지정
* index_col : '행 index' 가 되는 열 지정
* sep : 구분 할 문자
* names : 열 이름으로 사용할 문자열의 리스트
* skiprows : 생략할 행 (앞에서)
* skip_footer : 생략할 행 (뒤에서)
* parse_dates : 날짜 문자열을 datetime 형태로
* encoding : 예) 'utf-8'


```python
file_path = path + '/read_csv_sample.csv'
```

```python
df1 = pd.read_csv(file_path)
```

```python
df1
```

```python
df2 = pd.read_csv(file_path, header=None)
```

```python
df2
```

```python
df3 = pd.read_csv(file_path, index_col=None)
```

```python
df3
```

```python
df4 = pd.read_csv(file_path, index_col='c0')
```

```python
df4
```
