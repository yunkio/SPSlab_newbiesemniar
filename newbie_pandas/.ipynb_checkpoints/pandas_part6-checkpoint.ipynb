## 6-2. 함수 매핑

#### 예제 6-1. 시리즈의 원소에 apply() 적용

```python
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'fare']]
df['ten'] = 10
df.head()
```

```python
def add_10(n):
    return n + 10
def add_two_obj(a, b):
    return a + b

print(add_10(10))
print(add_two_obj(10, 10))
```

```python
sr1 = df['age'].apply(add_10)
sr1.head()
```

```python
sr2 = df['age'].apply(add_two_obj, b=10)
sr2.head()
```

```python
sr3 = df['age'].apply(lambda x: add_10(x))
sr3.head()
```

#### 예제 6-6. 데이터프레임에 pipe() 적용

```python
df = titanic_df[:]
```

```python
def missing_value(x):
    return x.isnull()

def missing_count(x):
    return missing_value(x).sum()

def total_number_missing(x):
    return missing_count(x).sum()
```

```python
result_df = df.pipe(missing_value)
result_df.head()
```

```python
result_df
```

```python
result_series = df.pipe(missing_count)
result_series
```

```python
type(result_series)
```

```python
result_value = df.pipe(total_number_missing)
result_value
```

```python
type(result_value)
```

## 6-3. 필터링

#### 예제 6-9. 불린 인덱싱

```python
mask1 = (titanic.age >= 10) & (titanic.age < 20)
df_teenage = titanic.loc[mask1, :]
df_teenage.head()
```

```python
mask2 = (titanic.age < 10) & (titanic.sex == 'female')
df_female_under10 = titanic.loc[mask2, :]
df_female_under10.head()
```

```python
mask3 = (titanic.age < 10) | (titanic.age >= 60)
df_under10_morethan60 = titanic.loc[mask3, ['age', 'sex', 'alone']]
df_under10_morethan60.head()
```

#### 예제 6-10. isin() 필터링

## 6-4. 데이터프레임 합치기

#### 예제 6-11. 데이터프레임 연결(concatenate)

```python
df1 = pd.DataFrame({'a' : ['a' + str(i) for i in range(4)],
                    'b' : ['b' + str(i) for i in range(4)],
                    'c' : ['c' + str(i) for i in range(4)]},
                    index = [0, 1, 2, 3])
df2 = pd.DataFrame({'a' : ['a' + str(i) for i in range(2, 6)],
                    'b' : ['b' + str(i) for i in range(2, 6)],
                    'c' : ['c' + str(i) for i in range(2, 6)],
                    'd' : ['d' + str(i) for i in range(2, 6)]},
                    index = [2, 3, 4, 5])
```

```python
df1
```

```python
df2
```

```python
result1 = pd.concat([df1, df2])
result1
```

```python
result2 = pd.concat([df1, df2], ignore_index=True)
result2
```

```python
result3 = pd.concat([df1, df2], axis=1)
result3
```

```python
result3_in = pd.concat([df1, df2], axis=1, join='inner')
result3_in
```

```python
sr1 = pd.Series(['e' + str(i) for i in range(4)], name = 'e')
sr2 = pd.Series(['f' + str(i) for i in range(3)], name = 'f', index = [3,4,5])
sr3 = pd.Series(['g' + str(i) for i in range(4)], name = 'g')
```

```python
result4 = pd.concat([df1, sr1], axis=1)
result4
```

```python
result5 = pd.concat([df2, sr2], axis=1, sort=True)
result5
```

```python
result6 = pd.concat([sr1, sr3], axis=1)
result6
```

```python
result7 = pd.concat([sr1, sr3], axis=0)
result7
```

## 6-14. 그룹 연산

#### 예제 6-14. 그룹 연산 - 분할

```python
df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]
print('승객 수:', len(df))
df.head()
```

```python
grouped = df.groupby(['class'])
grouped
```

```python
for key, group in grouped:
    print('* key :', key)
    print('* number :', len(group))
    print(group.head())
    print('----------------')
```

```python
average = grouped.mean()
average
```

```python
group3 = grouped.get_group('Third')
group3.head()
```

```python
grouped_two = df.groupby(['class', 'sex'])

for key, group in grouped_two:
    print('* key :', key)
    print('* number :', len(group))
    print(group.head())
    print('----------------')
```

```python
average_two = grouped_two.mean()
print(average_two)
print(type(average_two))
```

```python
group3f = grouped_two.get_group(('Third','female'))
group3f.head()
```

#### 예제 6-15. 데이터 집계

```python
grouped = df.groupby(['class'])
```

```python
std_all = grouped.std()
std_all
```

```python
std_fare = grouped.fare.std()
std_fare
```

```python
def min_max(x):
    return x.max() - x.min()

agg_minmax = grouped.agg(min_max)
agg_minmax.head()
```

```python
agg_all = grouped.agg(['min', 'max'])
agg_all.head()
```

```python
agg_sep = grouped.agg({'fare' : ['min', 'max'], 'age':'mean'})
agg_sep.head()
```
